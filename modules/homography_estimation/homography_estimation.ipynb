```python
import numpy as np
import cv2
from skimage import transform

# Sentido horario: sup. izq. -> sup. der. -> inf. der. -> inf. izq.
DIAGRAM_POINTS = [(170, 50), (288, 50), (288, 188), (170, 188)]

### Función para capturar clics del usuario sobre el frame
def get_user_points(image, window_name="Selecciona 4 puntos"):
    points = []

    def click_event(event, x, y, flags, param):
        if event == cv2.EVENT_LBUTTONDOWN and len(points) < 4:
            points.append((x, y))
            img_copy = image.copy()
            for pt in points:
                cv2.circle(img_copy, pt, 5, (0, 255, 0), -1)
            cv2.imshow(window_name, img_copy)
            if len(points) == 4:
                cv2.destroyAllWindows()

    cv2.imshow(window_name, image)
    cv2.setMouseCallback(window_name, click_event)
    cv2.waitKey(0)
    return points

def ordenar_puntos(puntos):
    puntos = np.array(puntos)

    idx_visual_ymin = np.argmax(puntos[:, 1])  # Más abajo en la imagen
    idx_visual_ymax = np.argmin(puntos[:, 1])  # Más arriba en la imagen

    lado_derecho = puntos[idx_visual_ymax][0] > puntos[idx_visual_ymin][0]

    print("Lado derecho:", lado_derecho)
    print("Puntos visuales:", puntos[idx_visual_ymin], puntos[idx_visual_ymax])

    idx_xmin = np.argmin(puntos[:, 0])
    idx_xmax = np.argmax(puntos[:, 0])
    idx_ymin = np.argmin(puntos[:, 1])  # visualmente más arriba (sup)
    idx_ymax = np.argmax(puntos[:, 1])  # visualmente más abajo (inf)

    if lado_derecho:
        # Canasta en lado derecho (visión del frame con canasta a la derecha)
        inf_izq = puntos[idx_xmin]
        sup_der = puntos[idx_xmax]
        inf_der = puntos[idx_ymax]
        sup_izq = puntos[idx_ymin]
    else:
        # Canasta en lado izquierdo
        inf_der = puntos[idx_xmax]
        sup_izq = puntos[idx_xmin]
        inf_izq = puntos[idx_ymax]
        sup_der = puntos[idx_ymin]

    return np.array([sup_izq, sup_der, inf_der, inf_izq])

### MAIN: Captura puntos, estima homografía, y muestra seguimiento del ratón

# 1. Carga primer frame del vídeo
video_path = "../../assets/clips/ClipS4.mp4"
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
cap.release()

if not ret:
    print("Error al leer el video.")
    exit()

# 2. Captura puntos sobre el frame
puntos_frame = get_user_points(frame.copy(), "Selecciona las 4 esquinas de la zona")

# 3. Ordena los puntos del frame
puntos_ordenados = ordenar_puntos(puntos_frame)

print("Puntos seleccionados del frame:", puntos_frame)
print("Puntos ordenados del frame:", puntos_ordenados)
print("Puntos del diagrama:", DIAGRAM_POINTS)

# 4. Estima la homografía
tform = transform.estimate_transform(
    'projective',
    np.array(puntos_ordenados, dtype=np.float32),
    np.array(DIAGRAM_POINTS, dtype=np.float32)
)

# 5. Carga el diagrama de cancha
diagrama = cv2.imread('../../assets/court_diagrams/fiba_white_court_half.png')

# 6. Configura evento para mostrar posición proyectada del ratón
def mouse_event(event, x, y, flags, param):
    if event == cv2.EVENT_MOUSEMOVE:
        punto_origen = np.array([[x, y]])
        punto_destino = param["tform"](punto_origen)
        imgtmp = param["image"].copy()
        cv2.circle(imgtmp, (int(punto_destino[0][0]), int(punto_destino[0][1])), 5, (0, 255, 0), -1)
        cv2.imshow(param["wname"], imgtmp)

# 7. Mostrar imágenes y activar seguimiento
cv2.imshow("Vista", frame)
cv2.imshow("Diagrama", diagrama)
cv2.setMouseCallback("Vista", mouse_event, {
    "tform": tform,
    "image": diagrama,
    "wname": "Diagrama"
})
cv2.waitKey(0)
cv2.destroyAllWindows()
```
